```python
def f(x):
    """
    2a. Einstein's Theory of Special Relativity f(x): E = mc^2 -> f(m) = E

    Args:
        x (int): mass (m) of particle in kg.

    Returns:
        e (int): Energy of the particle in Joules.
    """
    c = 3e8  # speed of light
    e = x * (c ** 2)

    return e
```

```python
# 2c. list of values y
list_y = [f(x) for x in list_x]
print(list_y)
```

```python
def logisticNext(x_n, r):
    """
    4a. Calculates the next element x_next (x_n+1) for a given x_n and r

    Args:
        x_n (float): the nth x element.
        r (float): r in the logistic iterator equation.

    Returns:
        x_next (float): (x_n+1) the element that foloows the nth x element.
    """

    x_next = r * x_n * (1 - x_n)
    return x_next
```

```python
# 4a. continued
x_n = 0.1
r = 1.5

print("First 20 elements:")
print(f"x_0: {x_n}")

for n in range(1, 20):
    x_n = logisticNext(x_n, r)
    print(f"x_{n}: {x_n}")
```

```python
def logisticSequence(x_0, r, n):
    """
    4b. Calculates the logisitic sequence up to element n and returns it as a list.

    Args:
        x_0 (float): the initial, 0th x element.
        r (float): r in the logistic iterator equation.
        n (int): the number of items in the sequence.

    Returns:
        sequence (array): the resulting logistic sequence as a list.
    """

    x_n = x_0
    sequence = [x_n]

    for i in range(1, n):
        x_n = logisticNext(x_n, r)
        sequence.append(x_n)

    return sequence
```

```python
# 4c.
from random import uniform

r = 2.5
n = 30
x_values = [x for x in range(0, n)]
min_limit = 0.0000001
max_limit = 1
x_0_1 = uniform(min_limit, max_limit)
x_0_2 = uniform(min_limit, max_limit)
x_0_3 = uniform(min_limit, max_limit)
x_0_4 = uniform(min_limit, max_limit)
x_0_5 = uniform(min_limit, max_limit)
sequence_1 = logisticSequence(x_0_1, r, n)
sequence_2 = logisticSequence(x_0_2, r, n)
sequence_3 = logisticSequence(x_0_3, r, n)
sequence_4 = logisticSequence(x_0_4, r, n)
sequence_5 = logisticSequence(x_0_5, r, n)

# Plot the sequences to verify variation
plt.plot(x_values, sequence_1, label=f"x_0: {x_0_1}")
plt.plot(x_values, sequence_2, label=f"x_0: {x_0_2}")
plt.plot(x_values, sequence_3, label=f"x_0: {x_0_3}")
plt.plot(x_values, sequence_4, label=f"x_0: {x_0_4}")
plt.plot(x_values, sequence_5, label=f"x_0: {x_0_5}")
plt.xlabel("x")
plt.ylabel("y (logisitic value)")
plt.title("Logistic Sequences for 5 Different Datasets with x_o within (1, 0)")
plt.legend()
plt.show()
```
